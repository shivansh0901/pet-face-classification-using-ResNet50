```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as image
import glob
import os
```

```python
images_fp = "./images/"
```

```python
image_names = [os.path.basename(file)for file in glob.glob(os.path.join(images_fp, "*.jpg"))]
```

```python
image_names
```

```python
labels = [name.split('_')[:-1:] for name in image_names]
```

```python
labels
```

```python
len(set(labels))
```

```python
def label_encode(labels):
    if labels == 'abyssinian': return 0
    if labels == "american shorthair": return 1
    if labels == "beagle": return 2
    if labels == "boxer": return 3
    if labels == "bulldog": return 4
    if labels == "chihuahua": return 5
    if labels == "corgi": return 6
    if labels == "dachshund": return 7
    if labels == "german shepherd": return 8
    if labels == "golden retriever": return 9
    if labels == "husky": return 10
    if labels == "labrador": return 11
    if labels == "maine coon": return 12
    if labels == "persian cat": return 13
    if labels == "pomeranian": return 14
    if labels == "pug": return 15
    if labels == "ragdoll cat": return 16
    if labels == "rottwiler": return 17
    if labels == "shiba inu": return 18
    if labels == "yorkshire terrier": return 19
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
features = []
labels = []
image_size = (224,224)

for name in image_names:
    label = ' '.join(name.split('_')[:-1:])
    label_encoded = label_encode(label)
    if label_encoded != None:
        img= load_img(os.path.join(images_fp, name))
        img= tf.image.resize_with_pad(img_to_array(img, dtype= 'uint8'), *image_size).numpy().astype('uint8')
        image = np.array(img)
        
        features.append(image)
        labels.append(label_encoded)
```

```python
len(labels)
```

```python
features_array=np.array(features)
labels_array = np.array(labels)
```

```python
features_array
```

```python
labels_array
```

```python
labels_one_hot = pd.get_dummies(labels_array)
```

```python
labels_one_hot
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(features_array, labels_one_hot, test_size= 0.2, random_state=10)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state= 35)
```

```python
from tensorflow.keras import layers, Input, Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input as ppi
from tensorflow.keras.layers import RandomRotation, RandomFlip, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy
```

```python
data_augmentation = Sequential([RandomFlip("horizontal"), RandomRotation(0.2)])
```

```python
predict_layers = Dense(20, activation= 'softmax')
```

```python
resnet_model = ResNet50(include_top=False, pooling= 'avg', weights= 'imagenet')
resnet_model.trainable= False
preprocess_input = ppi
```

```python
inputs = Input(shape=(224,224,3))
x= data_augmentation(inputs)
x= preprocess_input(x)
x=resnet_model(x, training= False)
x= Dropout(0.2)(x)
outputs= predict_layers(x)
model = Model(inputs, outputs)
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(), loss=CategoricalCrossentropy(), metrics=['accuracy'] )
```

```python
model_history = model.fit(x= X_train, y=y_train, validation_data=(X_val, y_val), epochs=1)
```

```python
model.evaluate(X_test, y_test)
```
